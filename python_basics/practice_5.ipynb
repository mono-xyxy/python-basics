# conditionals let your program to choose what to do based on condition


```python
age = 18
if age>=18:
    print("You can vote")
```

```python
marks = 45
if marks>=50:
    print("Pass")
else:
    print("Fail")
```

```python
score = 85
if score>=90:
    print("A Grade")
elif score >=75:
    print("B Grade")
elif score >=50:
    print("C Grade")
else:
    print("Fail")
```

```python
age = 20
has_id = True
if age>=18 and has_id:
    print("Entry allowed")
```

nested conditionals

```python
age = 20
citizen = True

if age>=18:
    if citizen:
        print("Eligible to vote")
    else:
        print("Not eligible to vote")
else:
    print("Too young")
```

```python
num = int(input("Enter a number: "))
if num%2==0 and num>0:
    print("Positive Even")
elif num>0 and num%2!=0:
    print("Positive Odd")
elif num<0 and num%2==0:
    print("Negative Even")
elif num <0 and num%2!=0:
    print("Negative Odd")
else:
    print("Zero")

```
