```python
nums = [3,1,5,2]
max_val = nums[0]
min_val = nums[0]
total = 0

for n in nums:
    total+=n

for n in nums:
    if n>max_val:
        max_val=n

for n in nums:
    if n<min_val:
        min_val=n

print(max_val,min_val,total)

```
