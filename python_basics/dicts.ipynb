- dict is key value mapping
- unordered
- mutable
- keys must be unique


```python
person={
    "name":"Ezzu",
    "age":23
}
person
```

```python
print(person["name"])
print(person.get("age"))
```

```python
# add / update
person["city"]="bangalore"
person["age"]=32
person
```

```python
# delete
person.pop("age")
del person["city"]
```

```python
# looping through keys of dictionaries
for key in person:
    print(key,person[key])
    # default key loop
```

```python
# loop through items()  
for k,v in person.items():
    print(k,v)
```

```python
# useful methods
print(person.keys())
print(person.values())
print(person.items())
print(len(person))
```

```python
# dictionaries used mainly for frequency counting
s="aabbc"
freq={}
for ch in s:
    freq[ch]=freq.get(ch,0)+1
print(freq)
```
