```python
# list methods
nums = [3,1,4,1]
len(nums)      # length
max(nums)      # max values
min(nums)      # min values
sum(nums)      # sum

nums.sort()    # sort ascending
nums.reverse() #reverse list
```

```python
for n in nums:
    print(n)
```

```python
nums=[1,2,3,4]
squares = [x**2 for x in nums]
print(squares)
```
