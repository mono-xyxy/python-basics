# Student marks analyser

```python
data=[
    ("alice",85),
    ("Bob",90),
    ("Alice",95),
    ("Bob",70),
    ("Charlie",60)
]
```

# Do the following:

- Create a dictionary with name → total marks
- Create another dictionary with name → average marks
- Find the student with highest average
- Filter students with average ≥ 80

```python
# Step-1 total marks + count
total={}
count={}

for name,marks in data:
    total[name] = total.get(name,0)+ marks
    count[name] = count.get(name,0)+1


avg={}
for name in total:
    avg[name] = total[name]/count[name]


max_avg=0
top_student=None

for name,value in avg.items():
    if value>max_avg:
        max_avg=value
        top_student=name

filtered = {}

for name,value in avg.items():
    if value>=80:
        filtered[name]=value

print("Total:", total)
print("Average:", avg)
print("Top:", top_student)
print("Filtered:", filtered)
```
