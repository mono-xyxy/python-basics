What is a set?
- unordered
- mutable
- stores unique values only
- No indexing


```python
s={1,2,3,3,4}
s
```

```python
# creating sets
s1={1,2,3}
s2 = set([1,2,2,3])
s3=set()
print(s1)
print(s2)
print(s3)
```

```python
# set core operations
s={1,2,3}
s.add(4)
s.remove(2)
s.discard(5)
s.pop()
s
```

```python
# membership check
print(3 in s)
```

```python
# set operations
a={1,2,3}
b = {3,4,5}
print(a | b)
print(a & b)
print(a-b)
print(a ^ b)
```

 remove duplicates

```python
nums = [1,2,2,3,4]
unique = list(set(nums))
unique
```

find common elements

```python
list1 = [1, 2, 3]
list2 = [2, 3, 4]
del list
common = list(set(list1) & set(list2))
print(common)
```

```python
# difference between list

only_in_1 = list(set(list1)-set(list2))
print(only_in_1)
```

```python
nums = [4,2,7,2,9,4,1]
unique = list(set(nums))
unique.sort()
print(unique)
```

```python
a = [1,2,3,4]
b = [3,4,5,6]

common = list(set(a) & set(b))
print(common)
only_in_a = list(set(a)-set(b))
print(only_in_a)
only_in_b = list(set(b)-set(a))
print(only_in_b)
```
