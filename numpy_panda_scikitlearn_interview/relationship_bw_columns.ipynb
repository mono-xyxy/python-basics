```python
import pandas as pd

data = {
    'age': [22, 25, 30, 35, 40, 45],
    'salary': [30000, 35000, 50000, 65000, 70000, 80000]
}

df = pd.DataFrame(data)

print(df)
```

```python
df[['age','salary']].corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x='age',y='salary',data=df)
plt.show()
```

```python
sns.heatmap(df.corr())
plt.show()
```
