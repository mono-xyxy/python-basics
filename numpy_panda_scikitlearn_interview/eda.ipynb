```python
import pandas as pd
import numpy as np

data = {
    'age': [25, 30, np.nan, 22],
    'salary': [50000, 60000, 55000, np.nan],
    'city': ['NYC', 'LA', 'NYC', 'SF']
}

df = pd.DataFrame(data)
```

```python
print(df.shape)
print(df.head())
print(df.dtypes)
print(df.describe())
print(df.isnull().sum())
```
