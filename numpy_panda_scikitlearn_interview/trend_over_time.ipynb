```python
import pandas as pd

data = {
    'date': ['2023-01-01', '2023-01-03', '2023-01-02', '2023-01-04'],
    'sales': [100, 300, 200, 400]
}

df = pd.DataFrame(data)

print(df)
```

```python
df['date']=pd.to_datetime(df['date'])
df
```

```python
df=df.sort_values('date')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['date'],df['sales'])
plt.show()
```
