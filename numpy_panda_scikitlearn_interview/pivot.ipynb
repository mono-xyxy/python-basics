```python
import pandas as pd

data = {
    'city': ['NYC', 'NYC', 'LA', 'LA'],
    'month': ['Jan', 'Feb', 'Jan', 'Feb'],
    'sales': [100, 150, 200, 250]
}

df = pd.DataFrame(data)

print(df)
```

```python
df_pivot=df.pivot_table(
    values="sales",
    index='city',
    columns='month',
    aggfunc='sum'
)
df_pivot=df_pivot.reset_index()
df_pivot
```

```python
df2 = pd.DataFrame({
    'id': [1, 2],
    'col1': [10, 20],
    'col2': [30, 40]
})
```

```python
df_melt=pd.melt(df2,id_vars=['id'],value_vars=['col1','col2'])
df_melt
```
