```python
import pandas as pd

data = {
    'salary': [50000, 52000, 51000, 1000000, 53000, 49000]  # 1000000 = outlier
}

df = pd.DataFrame(data)

print(df)
```

```python
Q1=df['salary'].quantile(0.25)
```

```python
Q3=df['salary'].quantile(0.75)
```

```python
IQR = Q3-Q1
```

```python
df=df[(df['salary']>=Q1-1.5*IQR) &( df['salary']<=Q3+1.5*IQR)]
```
