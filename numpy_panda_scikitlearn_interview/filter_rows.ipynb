```python
import pandas as pd
```

```python
data = {
    'age': [25, 35, 40, 28],
    'city': ['NYC', 'LA', 'NYC', 'SF'],
    'salary': [50000, 70000, 80000, 45000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df[df['age']>30]
```

```python
df[(df['age']>30) &(df['city']=='NYC')]
# Always use () with &
```

```python
df[(df['city']=="NYC") | df['city']=="LA"]
```

```python
df[df['city'].isin(['NYC','LA'])]
```

```python
df[df['age'].between(30,40)]
```
