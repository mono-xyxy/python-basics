```python
import pandas as pd

df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'id': [2, 3, 4],
    'salary': [50000, 60000, 70000]
})
```

```python
df=pd.merge(df1,df2,on='id',how='inner')
df
```

```python
df=pd.merge(df1,df2,on='id',how='left')
df
```

```python
df=pd.merge(df1,df2,on='id',how='outer')
df.reset_index(drop=True)
```
