```python
import pandas as pd

data = {
    'name': ['A', 'B', 'C', 'D', 'E'],
    'sales': [500, 700, 300, 900, 600],
    'city':['NYC','Germany','ABC','DEF','GHI']
}

df = pd.DataFrame(data)

print(df)
```

```python
df.sort_values('sales',ascending=False)
```

```python
df.nlargest(3,'sales')
```

```python
df.nsmallest(2,'sales')
```

```python
df.sort_values(['city','sales'],ascending=[True,False])
```

```python
df['rank']=df['sales'].rank(ascending=False)
```

```python
df.sort_values('sales',ascending=False).reset_index(drop=True)
```
