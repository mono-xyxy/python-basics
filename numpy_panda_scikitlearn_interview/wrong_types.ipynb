```python
import pandas as pd
```

```python
data = {
    'age': ['25', '30', 'not_available', '22'],
    'salary': ['50000', '60000', '55000', 'unknown'],
    'date': ['2023-01-01', '2023-02-15', 'invalid', '2023-03-10']
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
print(df.dtypes)
```

```python
df['age']=pd.to_numeric(df['age'],errors='coerce')
```

```python
df['salary']=pd.to_numeric(df['salary'],errors='coerce')
```

```python
df['date']=pd.to_datetime(df['date'],errors='coerce')
```
