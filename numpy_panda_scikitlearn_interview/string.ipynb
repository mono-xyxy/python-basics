```python
import pandas as pd
```

```python
data = {
    'name': [' Alice ', 'BOB', ' charlie!! ', 'DaVid  ']
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df['name']=df['name'].str.strip()
df['name']=df['name'].str.lower()
```

```python
df['name']=df['name'].str.replace('[^a-zA-Z0-9]','',regex=True)
```
